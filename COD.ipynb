```python
import pandas as pd
dataset=pd.read_csv("/content/cod.csv")
dataset.head()
```

```python
dataset.info()
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset.gamesPlayed,dataset.wins)
plt.title("COD Analysis")
plt.xlabel("GAme")
plt.ylabel("win")
plt.show();
```

```python
data=dataset.drop(columns=['name'])
```

```python
data.info()
```

```python
x=data.drop(columns=['wins'])
y=data['wins']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain,ytest=train_test_split(
    x,y,random_state=42,test_size=0.2
)
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_squared_error as mse
predicted=lr.predict(xtest)
error=mse(ytest,predicted)**.5
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
models=[RandomForestRegressor(),DecisionTreeRegressor(),LinearRegression()]

errors={}

for model in models:
  model.fit(xtrain,ytrain)
  name=type(model).__name__
  score=model.score(xtest,ytest)*100
  errors[name]=score
```

```python
errors
```

```python
import matplotlib.pyplot as plt
plt.bar(errors.keys(),errors.values())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
bars = plt.bar(errors.keys(), errors.values())

# Format bars
for bar in bars:
    bar.set_linewidth(0)
    bar.set_edgecolor('none')
    bar.set_zorder(3)

# Titles and labels
plt.title('Chart', fontsize=16, fontweight='bold')
plt.xlabel('Models', fontsize=12)
plt.ylabel('R values', fontsize=12)

# Background and grid
plt.gca().set_facecolor('#FFF0F5')
plt.grid(axis='y', linestyle='--', alpha=0.4, zorder=0)

# Remove top & right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(xtrain, ytrain)
print("Ridge score:", ridge.score(xtest, ytest))
```

```python
# Lasso Regression
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(xtrain, ytrain)
print("Lasso score:", lasso.score(xtest, ytest))
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
xtrain_poly = poly.fit_transform(xtrain)
xtest_poly = poly.transform(xtest)

poly_lr = LinearRegression()
poly_lr.fit(xtrain_poly, ytrain)
print("Polynomial Regression score:", poly_lr.score(xtest_poly, ytest))
```

```python
# K-Nearest Neighbors Regression
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(xtrain, ytrain)
print("KNN score:", knn.score(xtest, ytest))
```

```python
# Support Vector Regression
from sklearn.svm import SVR
svr = SVR()
svr.fit(xtrain, ytrain)
print("SVR score:", svr.score(xtest, ytest))
```

```python
# Gradient Boosting Regression
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr.fit(xtrain, ytrain)
print("Gradient Boosting score:", gbr.score(xtest, ytest))
```

```python
# XGBoost Regression
import xgboost as xgb
xgbr = xgb.XGBRegressor()
xgbr.fit(xtrain, ytrain)
print("XGBoost score:", xgbr.score(xtest, ytest))
```

```python
# LightGBM Regression
import lightgbm as lgb
lgbm = lgb.LGBMRegressor()
lgbm.fit(xtrain, ytrain)
print("LightGBM score:", lgbm.score(xtest, ytest))
```

```python
# Extra Trees Regressor
from sklearn.ensemble import ExtraTreesRegressor
etr = ExtraTreesRegressor()
etr.fit(xtrain, ytrain)
print("Extra Trees score:", etr.score(xtest, ytest))
```

```python
# AdaBoost Regressor
from sklearn.ensemble import AdaBoostRegressor
abr = AdaBoostRegressor()
abr.fit(xtrain, ytrain)
print("AdaBoost score:", abr.score(xtest, ytest))
```

```python
# Bagging Regressor
from sklearn.ensemble import BaggingRegressor
bgr = BaggingRegressor()
bgr.fit(xtrain, ytrain)
print("Bagging score:", bgr.score(xtest, ytest))
```
